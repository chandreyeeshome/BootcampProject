#Implementation of Machine Learning Models

---



---



LINEAR REGRESSION MODEL

---



```python
#Examining structure of data (processed)

df.head()
```

```python
#Splitting the dataframe as X(Years of dataframe) and Y(Salary)

X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
#Showing the values in X

X
```

```python
#Showing the values in Y

Y
```

```python
#Splitting the dataframe (into train and test sections)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
#LR Model --> lr (variable)

lr = LinearRegression()
```

```python
#Training the LR Model

lr.fit(X_train,Y_train)
```

```python
#Showing the X_test

X_test
```

```python
#Showing the Y_test

Y_test
```

```python
#Prediction

lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
#Evaluating the Model's performance (Checking accuracy)

Y_true = Y_test
Y_pred = lr.predict(X_test)

r2 = r2_score(Y_true, Y_pred)

print(f"R-squared Accuracy: {r2}")
```

```python
#Plotting the graph (regression line and the data points) to visualize the model's fit.

plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
#The Formula used by Linear Regression model: y = mx + c (where m is slope and c is y-intercept)

m = lr.coef_
c = lr.intercept_
```

```python
#Example --> y for x=10 i.e, predicted salary for 10 years of experience

y = m*10 + c

y
```

RANDOM FOREST REGRESSION


---







```python
#Examining structure of data

df.head()
```

```python
#Splitting the dataframe as X(Feature --> Years of Experience) and Y(Target --> Salary)

X = df.iloc[:,0:1]
Y = df.iloc[:,-1]
```

```python
#Splitting the dataframe (into train and test sections)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
#RF Model --> rf (variable)

rf = RandomForestRegressor(n_estimators=100, random_state=2)
```

```python
#Training the FR Model

rf.fit(X_train, Y_train)
```

```python
#Prediction

rf.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
#Evaluating the Model's performance (Checking accuracy)

Y_true = Y_test
Y_pred = rf.predict(X_test)

r2 = r2_score(Y_true,Y_pred)

print(f"R Squared Accuracy: {r2}")
```

```python
#Plotting the graph (predicted values against actual values) to visualize the model's fit.

#In a Random Forest Regression model, plotting the relationship between the features and the target variable
#can be challenging due to the ensemble nature of the model (RF is an ensemble of decision trees -> another ML Model).

#Because of this ensemble approach, the model's prediction surface in a Random Forest becomes much more complex and non-linear
#compared to a single decision tree or linear regression model.

#As a result, plotting a single regression line as done in a linear regression model is not meaningful and more complex in a Random Forest.
#Hence, we plot the predicted values against the actual values to visualize the model's performance.

plt.scatter(Y_true, Y_pred, color="red")
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.show()
```
